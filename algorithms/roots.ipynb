```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**3 - 10*x**2 + 5*x +80

def f_derivative(x):
    return 3*x**2 - 20*x + 5
```

# Graphical method

Finding the roots of an equation with the graphical method is a simple and intuitive way to solve it. The idea is to plot the function and find the points where it crosses the x-axis. These points are the roots of the equation.

```python
# Function to find the root of f(x) by graphical method. It should plot the function f(x) and the intercept points with x-axis
def graphical_method(f):
    x = np.linspace(-10, 10, 1000)
    y = f(x)
    plt.plot(x, y)
    plt.axhline(0, color="black", lw=0.5)
    plt.axvline(0, color="black", lw=0.5)
    plt.grid()
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.title("Graph of f(x) = x^3 - 10x^2 + 5x + 80")
    plt.show()


graphical_method(f)
```

# Segmentation method, incremental search

The idea is to divide the interval where the root is located into smaller intervals and check if the function changes sign in each interval. If the function changes sign, it means that there is a root in that interval.

```python
def incremental_segmentation_method(f, start, end, step):
    x = start
    while x <= end:
        if f(x) * f(x+step) <= 0:
            return x, x+step
        x += step
    return None

print('Interval of root of f(x) using incremental segmentation method:', incremental_segmentation_method(f, -10, 10, 0.01))
```

# Segmentation method, bisection

The idea is to divide the interval where the root is located into two smaller intervals and check in which interval the root is located. This process is repeated until the root is found with the desired precision.

```python
def bisection_method(f, start, end, tol):
    a = start
    b = end
    while b - a > tol:
        c = (a + b) / 2
        if f(a) * f(c) <= 0:
            b = c
        else:
            a = c
    return a, b

print('Interval of root of f(x) using bisection method:', bisection_method(f, -10, 10, 1e-6))
```

# Newton-Raphson method

The Newton-Raphson method is a numerical method for finding the roots of an equation. It is based on the idea of linear approximation of the function around a point. The method is iterative and converges to the root with a quadratic rate.

```python
def newton_raphson_method(f, f_derivative, start, tol):
    x = start
    while abs(f(x)) > tol:
        x = x - f(x) / f_derivative(x)
    return x

print('Root of f(x) using Newton-Raphson method:', newton_raphson_method(f, f_derivative, 0, 1e-6))
```
